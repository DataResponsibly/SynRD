```python
def plot_box_map(data, title, vmin=0, vmax=5, max_per_row=8):
    import matplotlib.pyplot as plt
    import seaborn as sb

    fig, ax = plt.subplots(figsize=(12, 8))
    corr = data.copy()
    # color map
    cmap = sb.color_palette("coolwarm_r", as_cmap=True)
    # plot heatmap
    ax = sb.heatmap(
        corr,
        annot=False,
        fmt=".0f",
        linewidths=5,
        cmap=cmap,
        vmin=vmin,
        vmax=vmax,
        square=True,
        cbar=False,
    )
    ax.set(xlabel="finding #")
    plt.yticks(plt.yticks()[0], rotation=0, fontsize=20)
    plt.xticks(plt.xticks()[0], fontsize=14)
    sb.set(font_scale=2)
    # plt.title(title, loc='left', fontsize=30)
    plt.show()
    return plt
```

```python
def plot_line_nested_epsilon_dict_error_bars(data):
    import seaborn as sns
    import pandas as pd

    g = sns.lineplot(
        data=pd.DataFrame(data),
        x="str_eps",
        y="value",
        hue="synth",
        style="synth",
        err_style="bars",
        ci=95,
        err_kws={"capsize": 5},
        legend=False,
    )
    g.set(xlabel=None, ylabel=None)
    return g
```
